Aqui, vamos criar nosso primeiro modelo simples usando TensorFlow e Keras. Vamos usar o Keras para construir uma rede neural sequencial, ou seja, com camadas empilhadas. Primeiro, adicionamos uma camada densa com 1 neurônio, que será responsável por gerar uma saída a partir de uma única entrada. Depois disso, compilamos a rede neural, definindo uma função de otimização e uma função de perda. A função de perda ajudará a medir os erros durante o treinamento, enquanto a função de otimização ajustará os pesos do modelo para reduzir esses erros e melhorar o aprendizado.

obs: o neuônio corresponde a quantidade de saída y.

obs: o modelo sequencial corresponde de várias camadas densas(o nosso é modelo sequncial de uma camada densa)


```python
model = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])
])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

Dados de entrada para treinamento do modelo, nessa etapa é usado dados de aprendizado, onde você deve fornecer pergunta x e resposta y, para assim o modelo aprender qual valor corresponde a seu resultado e generalizar essa lógica para demais casos.

```python

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

Captura de tela 2025-01-17 234504.png

Nesse segundo momemnto é passado para o treinamento os valores de entrada(dados que definimos) e quantidade de epócas que o treinamento irá estudar esses dados, visando uma boa curva de prendizado.

```python
#treinamento
model.fit(xs, ys, epochs= 500)
```

Na realização da predição, buscamos uma nova entrada(x) esperando que a resposta seja verdadeira com base no que o modelo estudou durante treinamento.

```python
# predição
input_data = np.array([20.0])
print(model.predict(input_data))
```

# criação de uma rede neural para prevê valores de imóveis, onde uma casa de 1 quarto custa 100k, uma casa de 2 quartos custa 150k e uma casa de 3 quartos custa 200k.

```python

model = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#dados treinamento
xy = np.array([0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
yx = np.array([ 50.0, 100.0, 150.0, 200.0, 250.0], dtype=float)
```

```python
model.fit(xy, yx, epochs=1000)
```

```python
input_data = np.array([1.0])
print(model.predict(input_data))
```
